```python
import pandas as pd
from pysus import SINAN
from pathlib import Path
```

```python
sinan = SINAN().load()
file = sinan.get_files(dis_code=["DENG"], year=2021)
sinan.describe(file[0])
```

```python
try:
    df = pd.read_parquet(Path.home() / "pysus" / "DENGBR21.parquet")
except Exception:
    df = sinan.download(sinan.get_files(dis_code=["DENG"], year=2021))
```

```python
df.head()
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
df[["DT_NOTIFIC", "SEM_NOT", "ID_MUNICIP", "CLASSI_FIN"]].head()
```

```python
df = df[df["CLASSI_FIN"].str.strip() != "5"]
df.shape
```

```python
df_dengue = df.groupby(["DT_NOTIFIC", "ID_MUNICIP"]).size().reset_index(name="case_count")
df_dengue["ID_MUNICIP"] = df_dengue["ID_MUNICIP"].astype(int)
print(df_dengue.shape)
df_dengue.head()
```

## IBGE

- https://www.ibge.gov.br/explica/codigos-dos-municipios.php (codigo municipios)

```python
df_ibge = pd.read_csv("data/RELATORIO_DTB_BRASIL_2024_MUNICIPIOS - DTB_Municípios.csv")
print(df_ibge.shape)
df_ibge.head()
```

```python
df_ibge.columns.tolist()
```

```python
df_ibge["id_municip"] = df_ibge["Código Município Completo"] // 10
# ISSO NAO FAZ SENTIDO KKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKK
```

## JUNTANDO

```python
df_merged = pd.merge(
    df_dengue,
    df_ibge,
    left_on="ID_MUNICIP",
    right_on="id_municip",
    how="left"
)
print(df_merged.shape)
df_merged.head()
```
